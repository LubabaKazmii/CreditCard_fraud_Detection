```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `data` is a Pandas DataFrame with columns 'TX_AMOUNT' and 'TX_FRAUD'
plt.figure(figsize=(10, 6))

# Pass the DataFrame and use column names for `x` and `hue`
sns.histplot(data=data, x='TX_AMOUNT', bins=50, kde=True, hue='TX_FRAUD')

plt.title("Transaction Amount Distribution")
plt.show()

```

```python
from sklearn.model_selection import train_test_split

X = data[['TX_AMOUNT', 'TX_TIME_SECONDS', 'TX_TIME_DAYS', 'CUSTOMER_ID', 'TERMINAL_ID']]
y = data['TX_FRAUD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

print("Training Accuracy:", model.score(X_train, y_train))
print("Test Accuracy:", model.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
import joblib

# Save the model
joblib.dump(model, "fraud_detection_model.pkl")

# Load the model
loaded_model = joblib.load("fraud_detection_model.pkl")

# Predict a new transaction
new_transaction = pd.DataFrame([{
    'TX_AMOUNT': 200,
    'TX_TIME_SECONDS': 50000,
    'TX_TIME_DAYS': 10,
    'CUSTOMER_ID': 1234,
    'TERMINAL_ID': 5678
}])

prediction = loaded_model.predict(new_transaction)
print("Prediction:", "Fraud" if prediction[0] == 1 else "Legitimate")
```
